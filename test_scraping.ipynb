```python
import os
import requests
import time
from apify_client import ApifyClient
from get_apify_limit import get_apify_limit
from ground_truth import GROUND_TRUTH, GROUND_TRUTH_BRANDS, GROUND_TRUTH_CREATORS

COLLAB_DIR = "/home/yash-sisodia/face-detection/collaborators_india/collaborators"

APIFY_KEYS = []
with open("apify_api_keys_yash.txt", 'r') as f:
    APIFY_KEYS = [line.strip() for line in f if line.strip()]

def download_image(display_url, shortcode, folder_path):
    os.makedirs(folder_path, exist_ok=True)
    image_filename = f"thumbnail_{shortcode}.jpg"
    image_path = os.path.join(folder_path, image_filename)

    try:
        resp = requests.get(display_url)
        if resp.status_code == 200:
            with open(image_path, "wb") as f:
                f.write(resp.content)
            print(f"Saved... {image_filename}")
        else:
            print(f"Failed... {image_filename}: HTTP {resp.status_code}")
    except Exception as e:
        print(f"Error downloading {image_filename}: {e}")

usernames = [d for d in os.listdir(COLLAB_DIR) if os.path.isdir(os.path.join(COLLAB_DIR, d))]

for idx, USERNAME in enumerate([u for u in usernames if u in GROUND_TRUTH], start=1):
    posts_dir = os.path.join(COLLAB_DIR, USERNAME, "latestPosts")
    igtv_dir = os.path.join(COLLAB_DIR, USERNAME, "latestIgtvVideos")

    if (os.path.exists(posts_dir) and os.listdir(posts_dir)) and (os.path.exists(igtv_dir) and os.listdir(igtv_dir)):
        print(f"\nSkipping... ({idx}) {USERNAME}, already scraped")
        continue

    print(f"\nFetching... ({idx}) {USERNAME}")

    current_apify_key = None
    for key in APIFY_KEYS:
        usage = get_apify_limit(key)
        if usage <= 4:
            current_apify_key = key
            break
    if current_apify_key is None:
        print("No key with usage limit more than $1 found")
        continue

    client = ApifyClient(current_apify_key)

    run_input = {
        "directUrls": [f"https://www.instagram.com/{USERNAME}/"],  
        "resultsLimit": 24,         # max posts to fetch
        "resultsType": "details"    # profile + posts + IGTV
    }
    try:
        run = client.actor("apify/instagram-scraper").call(run_input=run_input)
    except Exception as e:
        print(f"Failed to run scraper for {USERNAME}: {e}")
        continue

    items = list(client.dataset(run["defaultDatasetId"]).iterate_items())
    if not items:
        print(f"No data returned for {USERNAME}")
        continue

    download_count = 0
    for item in items:
        if "latestPosts" in item or "latestIgtvVideos" in item:
            content_types = {
                "latestIgtvVideos": item.get("latestIgtvVideos", []),
                "latestPosts": item.get("latestPosts", [])
            }
            for content_type, items_list in content_types.items():
                folder_path = os.path.join(COLLAB_DIR, USERNAME, content_type)
                for post in items_list:
                    if post.get("displayUrl") and post.get("shortCode"):
                        download_image(post["displayUrl"], post["shortCode"], folder_path)
                        download_count += 1

        elif "displayUrl" in item and "shortCode" in item:
            folder_path = os.path.join(COLLAB_DIR, USERNAME, "latestPosts")
            download_image(item["displayUrl"], item["shortCode"], folder_path)
            download_count += 1

    print(f"Downloaded... {download_count} images for {USERNAME}")
    time.sleep(3)
```
