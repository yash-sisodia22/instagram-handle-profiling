```python
import os
import numpy as np
from PIL import Image, ImageDraw
import insightface

COLLAB_DIR = "/home/yash-sisodia/face-detection/collaborators_india/collaborators"

SAMPLE_HANDLES = [
    "ellement.company",
    "ajiolife",
    "emprallofficial",
    "offduty.india",
    "himalaya_facecare",
    "lashupindia",
    "beromtnails",
    "fuaarkofficial",
    "staywrogn",
    "spykarofficial",
    "riseformen",
    "herbyangelofficial",
    "virgio.official",
    "kaybykatrina",
    "maccosmeticsindia",
    "suganda.co",
    "manyavar",
    "biotique_world",
    "82e.official",
    "uptownie101"
]
OUTPUT_DIR = "./bbox_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

model = insightface.app.FaceAnalysis(name="buffalo_l", providers=['CPUExecutionProvider'])
model.prepare(ctx_id=0, det_size=(640, 640))

def draw_bboxes(image_path, output_path):
    try:
        img = np.array(Image.open(image_path).convert("RGB"))
    except Exception as e:
        print(f"Could not open {image_path}: {e}")
        return False

    faces = model.get(img)
    if not faces:
        print(f"No faces detected in {image_path}")
        return False

    pil_img = Image.fromarray(img)
    draw = ImageDraw.Draw(pil_img)

    for face in faces:
        x1, y1, x2, y2 = face.bbox.astype(int)
        score = face.det_score
        draw.rectangle([x1, y1, x2, y2], outline="red", width=3)

    pil_img.save(output_path)
    print(f"Saved: {output_path}")
    return True

for username in SAMPLE_HANDLES:
    user_dir = os.path.join(COLLAB_DIR, username)

    if not os.path.isdir(user_dir):
        print(f"User {username} not found")
        continue

    user_output_dir = os.path.join(OUTPUT_DIR, username)
    os.makedirs(user_output_dir, exist_ok=True)

    image_paths = []
    profile_img = os.path.join(user_dir, "profile.jpg")
    if os.path.exists(profile_img):
        image_paths.append(profile_img)

    for folder in ["latestPosts", "latestIgtvVideos"]:
        folder_path = os.path.join(user_dir, folder)
        if os.path.exists(folder_path):
            for file in os.listdir(folder_path):
                if file.lower().endswith((".jpg", ".jpeg", ".png")):
                    image_paths.append(os.path.join(folder_path, file))

    for img_path in image_paths:
        img_name = os.path.basename(img_path)
        out_path = os.path.join(user_output_dir, img_name)
        draw_bboxes(img_path, out_path)

print("\nDOne.")
```
